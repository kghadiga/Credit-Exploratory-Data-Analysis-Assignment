As we can see that for all three continuous variables categorical columns AMT_INCOME_TOTAL, AMT_CREDIT & AMT_ANNUITY, there are outliers in huge numbers at 100th quartile.
So, to remove these outliers, we can use 1.5 IQR rule.

Lets create a outlier function which will return values that lies in lower range and upper range.

```python
def outlier_treatment(datacolumn):
 sorted(datacolumn)
 Q1,Q3 = np.percentile(datacolumn , [25,75])
 IQR = Q3 - Q1
 lower_range = Q1 - (1.5 * IQR)
 upper_range = Q3 + (1.5 * IQR)
 return lower_range,upper_range
```

#### Remove Outliers for AMT_INCOME_TOTAL Variable

Now, lets check lower range and upper range values using outlier_treatment function

```python
lower_range,upper_range = outlier_treatment(inp0.AMT_INCOME_TOTAL)
```

Now, let's check which column is outlier

```python
inp0[(inp0.AMT_INCOME_TOTAL < lower_range) | (inp0.AMT_INCOME_TOTAL > upper_range)]
```

Now, lets remove these outliers for AMT_INCOME_TOTAL Variable from the inp0.

```python
inp0.drop(inp0[(inp0.AMT_INCOME_TOTAL > upper_range) | (inp0.AMT_INCOME_TOTAL < lower_range) ].index, inplace=True)
```

Now, let's check if outliers are removed or not for AMT_INCOME_TOTAL variable

```python
#plot the boxplot again of AMT_INCOME_TOTAL variable to check if outliers are removed or not.
plt.figure(figsize=[12,6])
plt.ticklabel_format(style='plain')
sns.boxplot(inp0.AMT_INCOME_TOTAL)
plt.show()
```

```python
#Binning of AMT_INCOME_TOTAL variable.
q25 = inp0['AMT_INCOME_TOTAL'].quantile(0.25)
q60 = inp0['AMT_INCOME_TOTAL'].quantile(0.60)

q25,q60
```

```python
#Let's write a function to create Income Category groups for AMT_INCOME_TOTAL
def Income_Category(value):
    if value <= q25:
        return 'Low'
    elif value > q25 and value <= q60:
        return 'Medium'
    else:
        return 'High'
```
